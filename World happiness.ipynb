```python
# https://www.kaggle.com/unsdsn/world-happiness

import pandas as pd
import numpy as np

df2015 = pd.read_csv('data/2015.csv')
df2016 = pd.read_csv('data/2016.csv')
df2017 = pd.read_csv('data/2017.csv')
df2018 = pd.read_csv('data/2018.csv')
df2019 = pd.read_csv('data/2019.csv')
```

```python
# Add a year column

df2015['Year'] = 2015
df2016['Year'] = 2016
df2017['Year'] = 2017
df2018['Year'] = 2018
df2019['Year'] = 2019
```

```python
df2015['Social Support'] = np.nan
df2016['Social Support'] = np.nan
df2017['Social Support'] = np.nan
df2018['Dystopia Residual'] = np.nan
df2018['Family'] = np.nan
df2019['Dystopia Residual'] = np.nan
df2019['Family'] = np.nan
```

```python
# Drop columns

df2015 = df2015.drop(columns=['Standard Error', 'Region'])
df2016 = df2016.drop(columns=['Lower Confidence Interval', 'Upper Confidence Interval', 'Region'])
df2017 = df2017.drop(columns=['Whisker.high', 'Whisker.low'])
```

```python
# Update column names

df2015 = df2015.rename(columns={"Health (Life Expectancy)": "Healthy Life Expectancy", "Trust (Government Corruption)": "Perceptions of Corruption"})
df2016 = df2016.rename(columns={"Health (Life Expectancy)": "Healthy Life Expectancy", "Trust (Government Corruption)": "Perceptions of Corruption"})
df2017 = df2017.rename(columns={"Happiness.Score": "Happiness Score", "Happiness.Rank": "Happiness Rank", "Health..Life.Expectancy.": "Healthy Life Expectancy", "Trust..Government.Corruption.": "Perceptions of Corruption", "Economy..GDP.per.Capita.":"Economy (GDP per Capita)", "Dystopia.Residual": "Dystopia Residual"})
df2018 = df2018.rename(columns={"Score": "Happiness Score", "Overall rank": "Happiness Rank", "Country or region": "Country", "Freedom to make life choices": "Freedom", "GDP per capita": "Economy (GDP per Capita)", "Perceptions of corruption": "Perceptios of Corruption", "Healthy life expectancy": "Healthy Life Expectancy", "Social support": "Social Support"})
df2019 = df2019.rename(columns={"Score": "Happiness Score", "Overall rank": "Happiness Rank", "Country or region": "Country", "Freedom to make life choices": "Freedom", "GDP per capita": "Economy (GDP per Capita)", "Perceptions of corruption": "Perceptios of Corruption", "Healthy life expectancy": "Healthy Life Expectancy", "Social support": "Social Support"})
```

```python
# Combine datasets

df1516 = pd.merge(df2015, df2016, how="outer")
df1517 = pd.merge(df1516, df2017, how="outer")
df1518 = pd.merge(df1517, df2018, how="outer")
df1519 = pd.merge(df1518, df2019, how="outer")
```
